# RECOMMENDATION STEPS

```python
#1.Consider recommending songs to user_1  who is located at row number 0.
#2.Find the uclidan distance b/w User_1 and the remaining 999 users.
#3.Pick out top 5 users who all closest to user_1
#4.pick out the songs heard  by these top 5 users.
#5.Extract the songs which are maximum heard by 5 users and recommend the songs to user_1.
```

```python
def cal_dist(u1,u2):
    dist=pow(sum([pow(u1[x]-u2[x],2)for x in range(len(u1))]),0.5)
    return dist
```

```python
cal_dist(user_matrix[0],user_matrix[1])
```

```python
def distance_from_base_user(base_user,user_matrix):
    distance=[]
    
    for i in range(len(user_matrix)):
        if base_user!=i:
            distance.append(cal_dist(user_matrix[base_user],user_matrix[i]))
    return distance
```

```python
distance_from_base_user(0,user_matrix)
```

```python
distances=np.argsort(distance_from_base_user(0,user_matrix))[0:5]
```

```python
distances
```

```python
df.iloc[distances]
```

```python
def recommend_method_1(df,closest_users,n_songs):
    temp=df.iloc[closest_users]
    
    dict1=temp.max().to_dict()
    sorted_dict=sorted(dict1.items(),key=lambda keyvalue:(keyvalue[1],keyvalue[0]),reverse=True)[0:n_songs]
    return [x[0] for x in sorted_dict]
```

```python
recommend_method_1(df,distances,12)
```

# RECOMMENDATION METHOD_2

```python
#1 Create cluster of all the songs
#2 for a user to whom we are suppose to recommend a songs tag him as base user
#3 figure out the song commonly heard or frequently heard by the base user .
#4 check in which cluster does this song fall in
#5 calculate the neighbour distance between this song and other songs in the same cluster.
#6 pick out top 5 songs closest to this song
#7 recommmend these songs to the base user
```

```python
wcss={}
for k in range(1,40):
    kmeans=KMeans(n_clusters=k,random_state=10).fit(song_matrix)
    wcss[k]=kmeans.inertia_
```

```python
wcss
```

```python
plt.plot(wcss.keys(),wcss.values())
```

```python
## from the above curve we conclude that we will be taking 12 clusters
```

```python
def recommend_songs_methods_2(df,clusters,song_matrix,song_name,n_songs):
    
    ##making clusters of the song 
    kmeans=KMeans(n_clusters=clusters,random_state=40).fit(song_matrix)
    
    ## finding out the position of  favorite position 
    index_of_fav_song=df.columns.tolist().index(song_name)
    
    ##finding out the song vector of favourite song
    song_vector=song_matrix[index_of_fav_song]
    
    ##predicting the clusters in which songs are presemt
    all_songs_in_clusters=list(kmeans.predict(song_matrix))
    
    ##picking out other songs in the cluster where my favourite song is present in
    song_in_selected_cluster=[x for x in range(len(all_songs_in_clusters)) if all_songs_in_clusters[x]==kmeans.predict([song_vector])]
    
    ##picking out vectors of all the songs in the cluster
    song_cluster=song_matrix[song_in_selected_cluster]
    
    
    ##using nearrest neighnour to find the most closest song
    nn=NearestNeighbors(n_neighbors=n_songs)
    nn.fit(song_cluster)
    
    ## finding the recommended songs
    
    recommended_songs=nn.kneighbors([song_matrix[index_of_fav_song]])
    
    ## making a list of all the columns
    col_list=df.columns
    
    return[col_list[x] for x in recommended_songs[1][0]]
```

```python
recommend_songs_methods_2(df,12,song_matrix,'song_5',12)
```
